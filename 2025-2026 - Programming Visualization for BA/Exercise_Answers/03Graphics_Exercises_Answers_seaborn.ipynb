### a

```python
import pandas as pd
mtcars = pd.read_csv("mtcars.csv")
mtcars
```

### b
First produce a scatterplot of **wt** vs **hp**, then another scatterplot of **qsec** vs **hp**. 
Make the first one using seaborn and the second one using pandas.

```python
# Import seaborn and matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

# Create the first plot: a relplot using wt and hp as x and y
plot_qsec_vs_hp = sns.relplot(data = mtcars, x = "wt", y = "hp")

# Use plt.show() to show only the plot and not the additional textual output
plt.show()

# Use pandas to create the second plot
mtcars.plot.scatter("qsec", "hp")

plt.show()
```

### c
Produce histograms of **gear** and **wt**. Play with the number of bins for the latter one to get a more complete picture of the distribution.


```python
sns.histplot(mtcars, x = "gear")
plt.show()

sns.histplot(mtcars, x = "disp", bins = 30)
plt.show()
```

### d
Produce a density plot of **wt**, do this both using pandas and using seaborn.

```python
# Density plot using pandas
mtcars["wt"].plot(kind='density')

# Density plot using seaborn
sns.displot(mtcars, x = "wt", kind = "kde")
plt.show()
```
