### Five basic functions

#### 4. assign()

Example: Create a column giving the gain in delay, and a column
giving the gain per minute

```python
flights_selection = flights.loc[:, 'year':'air_time']
print(flights_selection.head(2))
```

```python
# Here we create a new column gain which is the difference between dep_delay and arr_delay
flights_gain = flights_selection.assign(gain = flights_selection['dep_delay']-flights_selection['arr_delay'])
print(flights_gain.head(2))
```

the second assign(gain_per_minute) can use the first assign(gain)

```python
# Now we create a new column gain_per_minute which is the gain divided by air_time
flights_gain = flights_gain.assign(gain_per_minute = flights_gain['gain']/flights_gain['air_time'])
print(flights_gain.head(2))
```

```python
# Change gain_per_minute to gain_per_hour
flights_gain = flights_gain.assign(gain_per_hour = flights_gain['gain_per_minute']*60)
print(flights_gain.head(2))
```

We don't go over this in class but: .expanding() creates an expanding window that grows with each row, allowing you to calculate cumulative statistics.

```python
# Calculates the mean of gain_per_hour over all rows up to the current row
flights_gain = flights_gain.assign(cumulative_gain = flights_gain['gain_per_hour'].expanding().mean())
print(flights_gain.head())
```

### Five basic functions

#### 5. groupby() and agg()

Example: Get the mean of dep_delay

```python
# Calculate only the mean
flights_summary = flights["dep_delay"].mean()

print(flights_summary)
```

Not so interesting. Better: Get the mean of dep_delay by day

```python
# Group flights by month and by day. Then calculate the mean departure delay per day.
flights_grouped = flights.groupby(['month', 'day'])

# This part is necessary to see the results
flights_grouped["dep_delay"].mean()
```

```python
# Shows the first row from each group. (each day in each month = 365 rows). Not helpful instead use descriptive functions
flights_grouped[['dep_delay']].head(1)
```

.agg() (aggregate) applies one or multiple functions to grouped data, making it easy to calculate descriptive statistics.

```python
# Calculate multiple functions simultaneously
#flights_grouped.agg({'dep_delay':['describe']})
flights_grouped.agg({'dep_delay':['mean', 'first', len]})
```

You can also change the column names

```python
# Assign column names
flights_grouped.agg(delay = ('dep_delay', 'mean'), first_delay = ('dep_delay', 'first'), nr_flights = ('dep_delay', len))
```

### Combine multiple functions. 

Various functions can also be used together with groupby,
Example: Find, for each day, the 9 most delayed flights

```python
# Find the most delayed flights.

flights_most_delay = flights.sort_values('arr_delay', ascending=False).groupby(['month', 'day']).head(9)
flights_most_delay
```

assign() and query() can also be used together with groupby
Example: Calculate, by destination, how much a flight adds to the
total delay

```python
# Filter flights with an arrival delay, then calculate the proportional delay
flights_delays = flights.query("arr_delay > 0")

flights_delays = flights_delays.assign(prop_delay = flights_delays['arr_delay']/sum(flights_delays['arr_delay']) * 100).groupby('dest').sum()
flights_delays["prop_delay"].sort_values(ascending=False)
```

### Pipes, cleaner coding

Having to save every step to a data.frame in your environment is
not very eﬃcient
−→ ’In between’-results are not of interest
Therefore, use pipes (adding functions after another) instead

```python
# Pipes or method chaining. Use parentheses to wrap the entire statement
flights_summary = (flights
                        .groupby(['month', 'day'])
                        .agg(delay = ('dep_delay', 'mean'), 
                             first_delay = ('dep_delay', 'first'), 
                             nr_flights = ('dep_delay', len))
                  )
flights_summary.head()                   
                   
```

You do not have to repeat the data you work with and do not have
to save every ’In between’-result!

```python
# Combine query, filter, assign, groupby, and agg
flights_gains = (flights 
                    .query("month == 3 & day == 13") 
                    .filter(['carrier', 'origin', 'dest', 'dep_delay', 'arr_delay'])
                    .assign(gain = flights.dep_delay - flights.arr_delay)
                    .groupby('dest')
                    .agg(mean_gain = ('gain', 'mean'), max_gain = ('gain', 'max'), nr_flights = ('gain', len))
                )
flights_gains
```

### Eﬃcient coding
Applying the same function to a set of columns of the data frame. Example: Find the average for dep_delay, arr_delay and
air_time

```python
# Calculate the mean function for multiple columns. Typing 'mean' for each column is not very efficient.
flights_across = flights_grouped.agg({'dep_delay':'mean', 'arr_delay':'mean', 'air_time':'mean'})
flights_across.head()
```

Having to copy the same code (i.e. the ’mean’ function) is not very
eﬃcient.
Better to first retrieve the desired columns from the groupby object
and then call the mean function only once.



```python
# More efficient way of calculating the mean for multiple columns
flights_across = flights_grouped[['dep_delay', 'arr_delay', 'air_time']].agg('mean')
flights_across.head()
```

```python
# Or
flights_test = flights_grouped[['dep_delay', 'arr_delay', 'air_time']].mean()
flights_test.head()
```

Several functions can be used for extracting columns, str.isalpha() Finds column names that are purely alphabetic (no numbers/symbols) .nunique() gets all the unique values from those columns

```python
# Find the number of distinct values for all character variables in flights
flights_across = flights[flights.columns[flights.columns.str.isalpha()]].nunique()
print(flights_across)
```

str.startswith() to all columns that start with, the parameter in the ()

```python
# Find the number of distinct values for all columns starting with 'dep_'
flights_across2 = flights[flights.columns[(flights.columns).str.startswith('dep_')]].nunique()
print(flights_across2)
```

flights.dtypes == object, get all columns of the datatype object specified

```python
# Find the number of distinct values for all alphabetic columns in flights. (object dtype is equal to a string)
flights_across3 = flights[flights.columns[flights.dtypes == object]].nunique()
print(flights_across3)
```

Several function inputs can be used together, e.g. mean and median
Example Find the mean and median, by day, for dep_delay,
arr_delay and air_time

```python
# Calculate both the mean and the median per day wihtout pipes
flights_grouped = flights.groupby(['month', 'day'])
flights_across = flights_grouped[['dep_delay', 'arr_delay', 'air_time']].agg(['mean', 'median'])
print(flights_across.head(10))
                 
```

```python
# Example Find the mean and median, by day, for dep_delay, arr_delay and air_time using pipes
flights_across = (flights
                  .groupby(['month', 'day'])
                  [['dep_delay', 'arr_delay', 'air_time']]
                  .agg(['mean', 'median'])
                 )
print(flights_across.head(10))
```

```python
# Untidy data (cases and population are in one column)
table2 = pd.read_csv("table2.csv")
print(table2)
```

Above data is untidy, since variables cases and population are together

```python
# Often data is spread over mutliple columns
table4a = pd.read_csv("table4a.csv")
print(table4a)
```

Columns ‘2019‘ and ‘2020‘ are actually values of the variable year. We can gather data spread over columns using pd.melt

pd.melt() transforms data from wide format to long format - it "melts" columns into rows. Here's how it works:

```python
# Gather data spread over columns pd.melt() converts multiple columns into two columns - one for the variable names and one for their values.
table_gathered = pd.melt(table4a, 
        id_vars=['country'],           # Columns to keep as identifiers
        value_vars=['1999', '2000'],   # Columns to melt/unpivot
        var_name='year',               # Name for the new variable column
        value_name='cases')            # Name for the new value column
print(table_gathered)

```

```python
print(table2)
```

Often, more than one variable is gathered in one column -> We can spread data gahtered in the column using pd.pivot

pd.pivot() transforms data from long format to wide format - it "spreads" rows into columns. It's the opposite of melt().

```python
# Spread data gathered in one column
table_spread = pd.pivot(table2, 
         index=['country', 'year'],    # Rows to keep as identifiers
         columns='type',               # Column to spread into new columns
         values='count')               # Values to fill the new columns
print(table_spread)
```

### melt(): Many columns → Few columns (wide to long)
### pivot(): Few columns → Many columns (long to wide)
