# Graphics

You can use https://www.data-to-viz.com to see an in-depth look in all the different ways you can show your data in plots.

```python
# Import pandas
import pandas as pd
```

```python
storms = pd.read_csv("storms.csv")
storms[["pressure"]].plot()
```

pd.plotting.scatter_matrix creates a matrix of scatter plots showing relationships between multiple variables at once.

1. Creates a grid where each variable is plotted against every other variable
2. Diagonal: Shows histograms of each individual variable
3. Off-diagonal: Shows scatter plots between pairs of variables


```python
pd.plotting.scatter_matrix(storms[["status", "category", 
                               "wind", "pressure"]])
```

```python
import matplotlib.pyplot as plt

plt.plot(storms["pressure"])
```

```python
import seaborn as sns
```

sns.relplot() is called a relplot because it's designed to show relationships between variables, not because of the specific plot type it creates. The default relplot is a scatterplot.

```python
sns.relplot(data = storms, x = "wind", y = "pressure")
```

```python
# Create histogram
sns.histplot(data = storms, x = "wind")
```

Instead of integers you can also use strings in the bins parameter: 
1. 'auto': Good default, lets numpy choose the best method
2. 'fd': Good for large datasets, robust to outliers
3. 'scott': Good for normal distributions
4. 'sqrt': Simple rule, works well for small datasets

```python
# Adjusting the bin width 
sns.histplot(data = storms, x = "wind", bins = 15)
```

Using stat="density" we normalize the data shown

```python
sns.histplot(data = storms, x = "wind", bins = 15, stat = "density")
```

sns.displot() is seaborn's figure-level function for distribution.  Kind is the type of plot:  "kde" = Kernel Density Estimate - smooth density curve

```python
sns.displot(storms, x = "wind", kind = "kde")
```

```python
storms[["wind"]].plot(kind='density')
```

```python
import statsmodels.api as sm
```

1. Sample Quantiles: Your actual data
2. Theoretical quantiles: Expected distribution if it followed the line perfectly


```python
fig = sm.qqplot(storms['wind'])
plt.show()

# Add theoretical line
#fig = sm.qqplot(storms['wind'], line = "45", fit = True)
#plt.show()
```

Boxplots show specific quantiles in parts:
1. Bottom whisker: Minimum (or 1.5IQR below Q1)
2. Bottom of box: 25th percentile (Q1)
3. Middle line: 50th percentile (Q2 = Median)
4. Top of box: 75th percentile (Q3)
5. Top whisker: Maximum (or 1.5IQR above Q3)
6. Dots: Outliers beyond whiskers


```python
sns.boxplot(data = storms, y = "wind")
plt.show()
```

```python
sns.boxplot(data = storms, y = "wind", x = "month")
plt.show()
```

```python
sns.countplot(data = storms, x = 'status')
plt.show()
```
